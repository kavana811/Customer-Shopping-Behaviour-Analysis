```python
##filling median for null values from median review rating with in each category median 
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#changing column name where space is there inserting underscore
df.columns =df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
#changing purchase column name 
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#creating a column sperate age group based on age
labels = ['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#createing column purchase_frequency_days
frequency_mapping = {'Fortnightly':14,
                     'Weekly':7,
                     'Monthly':30,
                     'Quarterly':90,
                     'Bi-Weekly':14,
                     'Annually':365,
                     'Every 3 Months':90
                    }
df['purchase_frequency-days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency-days','frequency_of_purchases']]
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
#droping not needed column
df=df.drop('promo_code_used', axis=1)
df.columns
```
